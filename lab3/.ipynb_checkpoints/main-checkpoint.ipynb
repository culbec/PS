```python
from random import sample, randrange
from math import comb
from matplotlib.pyplot import bar, hist, grid, show, legend
from scipy.stats import binom
from numpy import arange
from collections import Counter


def bile(noSimulations: int):
    noValidBA = 0
    noValidABarat = 0

    colors = ['red', 'blue', 'green']

    for _ in range(noSimulations):
        item = sample(colors, k=3, counts=[5, 3, 2])

        if len(set(item)) == 1 and item[0] == 'red':
            noValidBA = noValidBA + 1
        
        valid = True
        for color in item:
            if color == 'red':
                valid = False
                break
        if valid:
            noValidABarat = noValidABarat + 1

    probA = 1 - noValidABarat / noSimulations
    probBA = noValidBA / noSimulations
    probTeoretica = comb(5, 3) / (comb(10, 3) - comb(5, 3))
    print("Probabilitatea prin simulari repetate pentru nrSimulari=", noSimulations, ": ",
          float(probBA) / probA)
    print("Probabilitatea teoretica: ", probTeoretica)
    print("\n")


def zar():
    # afisam probabilitatea simulata pentru fiecare numar
    data = [randrange(1, 7) for _ in range(500)]
    bin_edges = [k + 0.5 for k in range(7)]
    hist(data, bin_edges, density=True, rwidth=0.9, color='green', edgecolor='black', alpha=0.6,
         label='probabilitate simulata zar')

    # pentru toate numerele exista aceeasi probabilitate de nimerire
    distribution = dict([(i, 1 / 6) for i in range(1, 7)])
    bar(distribution.keys(), distribution.values(), width=0.85, color='red', edgecolor='black', alpha=0.6,
        label='probabilitatea teoretica zar')

    legend(loc='lower left')
    grid()
    show()


def probX():
    noExtrageri = 5
    prob_1 = 6 / 10

    X = binom.rvs(noExtrageri, prob_1, size=1000)
    print('1000 de valori pentru X')
    print(X)

    bin_edges = [k + 0.5 for k in range(7)]
    hist(X, bin_edges, density=True, color='green', alpha=0.6, label='Frecvente relative')

    XTeoretic = arange(0, 6)
    pmf_values = binom.pmf(XTeoretic, noExtrageri, prob_1)
    bar(XTeoretic, pmf_values, color='red', alpha=0.6, label='Probabilitatea teoretica')

    legend(loc='lower left')
    grid()
    show()

    dictValues = dict((i, list(X).count(i)) for i in range(3, 6))
    sumValues = sum(dictValues.values())
    probabilitateSimulata = float(sumValues) / 1000
    probabilitate = binom.cdf(5, noExtrageri, prob_1) - binom.cdf(2, noExtrageri, prob_1)
    print('Probabilitate simulata P(2 < X <= 5) este: ', probabilitateSimulata)
    print('Probabilitatea teoretica P(2 < X <= 5) este: ', probabilitate)

def generareSume():
    return sum(randrange(1, 7) for _ in range(3))

def sumeZar(noSimulations: int):
    
    sume = [generareSume() for _ in range(noSimulations)]
    frecvSume = Counter(sume)
    mostFreq = frecvSume.most_common(3)
    print("Cele mai comune sume: ", mostFreq)
    
    bin_range = [i for i in range(2, 20)]
    hist(sume, bin_range, density=True, color='green', alpha=0.6, edgecolor='black', label="Probabilitate simulata")
    
    probTeoretice = []
    
    for suma in range(3, 19):
        noValid = 0
        for i in range(1, 7):
            for j in range(1, 7):
                for k in range(1, 7):
                    if i + j + k == suma:
                        noValid = noValid + 1
        probTeoretice.append(noValid / 216)
    
    bar(range(3, 19), probTeoretice, color='red', alpha=0.6, edgecolor='black', label='Probabilitati teoretice')
                        
    legend(loc='lower left')
    grid()
    show()

def main():
    bile(20000)
    bile(10000000)
    zar()
    probX()
    sumeZar(10000)

main()
```
