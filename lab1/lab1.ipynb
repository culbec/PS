```python
from random import sample
from math import factorial
from itertools import permutations

cuvant = "abecedar"

# a) lista cu permutarile cuvantului dat
permCuv = list(''.join(cuv) for cuv in permutations(cuvant))
print(permCuv)

# b) permutari totale
print("Permutari totale ale cuvantului " + cuvant + " sunt: " + str(factorial(len(cuvant))))

# c) permutare random
print("Permutare random a cuvantului " + cuvant + ": " + sample(permCuv, 1)[0])
```

```python
from random import sample
from math import perm, comb
from itertools import permutations, combinations

def aranjamente(cuvant: str, size: int, numar_total=False, aleator=False):
    if numar_total is True:
        print("Aranjamentele totale sunt: " + str(perm(len(cuvant), size)))
        return
    
    aranj = list(''.join(cuv) for cuv in permutations(cuvant, size))
    
    if aleator is True:
        print("Aranjament random: " + str(sample(aranj, 1)[0]))
        return
    
    print("Aranjamentele cuvantului " + cuvant + " sunt: ")
    print(aranj)
    
aranjamente('word', 2)
aranjamente('word', 2, numar_total=True)
aranjamente('word', 2, aleator=True)

print()

def combinari(cuvant: str, size: int, numar_total=False, aleator=False):
    if numar_total is True:
        print("Combinarile totale sunt: " + str(comb(len(cuvant), size)))
        return
    
    combs = list(''.join(cuv) for cuv in combinations(cuvant, size))
    
    if aleator is True:
        print("Combinare random: " + str(sample(combs, 1)[0]))
        return
    
    print("Aranjamentele cuvantului " + cuvant + " sunt: ")
    print(combs)
    
combinari('word', 2)
combinari('word', 2, numar_total=True)
combinari('word', 2, aleator=True)
    
```
